```python
import pandas as pd
import numpy as np
import requests
import getpass
from urllib import parse
import re
```

```python
medecine_name = "paracétamol"
params = {"query": medecine_name}
url_params = parse.urlencode(params)
medicament_base_url = "https://www.open-medicaments.fr/api/v1/medicaments"
medecines_url = medicament_base_url + "?" + url_params
medecines_url
```

```python

response = requests.get(medecines_url)
medecines = response.json()
for medecine in medecines:
    match = re.search("([0-9]+) (mg|g), (.*)", medecine['denomination'])
    dose = match.group(1)
    qte = match.group(2)
    form = match.group(3)
    print(dose, qte, form)
    medecine_info_url = medicament_base_url + "/" + medecine['codeCIS']
```

```python
df = pd.DataFrame(medecines)
regex = re.compile("([\D]*)([0-9]+) (mg|g), (.*)")
ds = df['denomination'].str.extract(regex)
ds['mul'] = ds[2].map({'mg': 1000, 'g': 1})
ds
```
